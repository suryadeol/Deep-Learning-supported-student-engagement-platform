```python
from keras.models import load_model
from time import sleep
from tensorflow.keras.preprocessing.image import img_to_array

from keras.preprocessing import image
import cv2
import numpy as np

#face_classifier = cv2.CascadeClassifier(r'haarcascade_frontalface_default.xml')
face_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
classifier =load_model(r'model.h5')
person_model = load_model('personal.h5')
```

```python
emotion_labels = ['Angry','Disgust','Fear','Happy','Neutral', 'Sad', 'Surprise']
#emotion_labels = ['irritated', 'uninterested', 'apprehensive', 'enthusiastic' , 'observant', 'distracted', 'curious']
members_label = ['ibraheem', 'sai krishna', 'surya']

cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    labels = []
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray)

    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,255),2)
        roi_gray = gray[y:y+h,x:x+w]
        

        
        
        
        roi_gray = cv2.resize(roi_gray,(48,48),interpolation=cv2.INTER_AREA)

    

        if np.sum([roi_gray])!=0:
            roi = roi_gray.astype('float')/255.0
            roi = img_to_array(roi)
            roi = np.expand_dims(roi,axis=0)
            
            
            #Predict person identification label using the model
            person_label_index = tf.argmax(person_model.predict(roi)[0]).numpy()
            #person_label = str(person_label_index)  
            
            

            

            prediction = classifier.predict(roi)[0]
            label=emotion_labels[prediction.argmax()]
            label_position = (x,y)
            cv2.putText(frame,label,label_position,cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
            cv2.putText(frame, members_label[person_label_index] ,label_position,cv2.FONT_HERSHEY_SIMPLEX,1,(0,25,0),2)
            
        else:
            cv2.putText(frame,'No Faces',(30,80),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
    cv2.imshow('Emotion Detector',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Define engagement labels


emotion_labels = ['Angry','Disgust','Fear','Happy','Neutral', 'Sad', 'Surprise']
# Initialize variables to count the occurrence of each engagement label for each person
engagement_counts = {str(label): 0 for label in range(3)}  # Adjust NUM_PERSONS based on your person model output
engage = {str(label): "" for label in range(3)}  # Dictionary to store the most frequent engagement label for each person

# Function to detect faces, identify persons, and estimate engagement levels
def detect_faces_identify_persons_estimate_engagement(frame):
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        
        face_roi = cv2.resize(frame[y:y+h, x:x+w], (256, 256))  # No normalization
        
        #for person identification
        #face_identify = cv2.cvtColor(face_roi, cv2.COLOR_BGR2RGB)
        #face_identify = face_identify/255.0
        #face_identify = tf.expand_dims(face_identify,axis=0)
        
        #for engagement
        face_roi = cv2.cvtColor(face_roi, cv2.COLOR_RGB2GRAY)
        face_roi = face_roi.reshape(256,256,1)
        face_roi = tf.image.resize(face_roi, (48, 48))
        face_roi = tf.expand_dims(face_roi, axis=0)
       
        
        #Predict person identification label using the model
        #person_label_index = tf.argmax(person_model.predict(face_identify)[0]).numpy()
        #person_label = str(person_label_index)  # Convert to string
    
        # Predict engagement level using the student model
        predictions = model.predict(face_roi)
        engagement_label_index = tf.argmax(predictions[0]).numpy()
        engagement_label = engagement_labels[engagement_label_index]

        # Draw rectangle around the face and display engagement label and person label
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        #cv2.putText(frame, f"Person: {person_label}", (x, y-30), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)
        cv2.putText(frame, f"Engagement: {engagement_label}", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)

        # Update engagement counts and most frequent engagement label for each person
        if person_label in engagement_counts:
            engagement_counts[person_label] += 1
            if engagement_counts[person_label] >= max(engagement_counts.values()):
                engage[person_label] = engagement_label 

    return frame 

# Capture video from webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    
    if not ret:
        break

    # Perform face detection, person identification, and engagement estimation
    frame = detect_faces_identify_persons_estimate_engagement(frame)

    # Display the processed frame with engagement labels
    cv2.imshow('Face Detection, Person Identification, and Engagement Estimation', frame)

    # Exit the loop if the user presses 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()

# Print the most frequent engagement label for each person
for person, label in engage.items():
    print(f"Most frequent engagement label for person {person}: {label}")'
```
